# Airplane Crashes since 1908

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd ##load data set and first look
```

```python
df = pd.read_csv("Data/AirplaneCrashes.csv")
df
```

```python
#Overview
print(df.head(3))

#Which Colums are contained?
print(list(df.columns))
```

```python
####Calculate Survivors Plane, People, Survivors total, total deaths
df['surv_plane']= df.Aboard - df.Fatalities

df['people'] = df.Aboard + df.Ground

df['surv_tot'] = df.people -df.Fatalities -df.Ground

df['death_tot'] = df.people - df.surv_tot	

###Checking some Values
df.loc[df['Aboard'] == 0]
```

```python
# Most crashing flight number
df.loc[df["Flight #"] != "-"].groupby("Flight #").count().sort_values("Date", ascending=False)
```

```python
# Most deadliest flight number
df.loc[df["Flight #"] != "-"].groupby("Flight #").sum().sort_values("death_tot", ascending=False)
```

```python
#Are the two most deadliest flight numbers (11,175) from 9/11?

df.loc[df["Flight #"] != "-"].loc[df["Flight #"].isin(['11','175'])]
```

```python
#Flights with max number death

id_max_casualties = df['death_tot'].idxmax()
df.iloc[id_max_casualties]

df.iloc[id_max_casualties].Summary
```

```python
# Number of Flights which got hijacked

len(df[df['Summary'].str.contains("hijack" , na = False)])
```

```python
# Hijacked Flights

hi = df[df['Summary'].str.contains("hijack" , na = False)]
                                                                          
hi['Date'] = pd.to_datetime(hi['Date'])

hi['Year'] = hi['Date'].map(lambda x: x.year)

hi
```

```python
#plt.plot_date(hi['Date'], hi['Flight #'].count())
hi ##['Date']

hi.groupby("Year").count()['Date'].plot.bar()
```

```python
#Operators with most deaths
df.loc[df["Operator"] != "-"].groupby("Operator").sum().sort_values("death_tot", ascending=False)
```

```python
#Operators with most total survivors
df.loc[df["Operator"] != "-"].groupby("Operator").sum().sort_values("surv_tot", ascending=False)
```
